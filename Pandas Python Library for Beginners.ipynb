## Task 1 : Creating a Series 

### Using an Array

Create a Pandas Series, named s1, containing the following:
* Integer = 1
* Nan value
* Character = a

```python
s1 = np.array([1,np.nan,"a"])
s1 = pd.Series(s1)
s1
```

### Using a List

Create a Pandas Series, named s2.
Use a list, named list2, containing the following:
* Integer = 2
* Character = b
* String = Numpy 

```python
list2 = ["2","b","Numpy"]
s2 = pd.Series(list2)
s2
```

### Using a Dictionary

Create a Pandas Series, named s3, containing the following:
* "KEYS" = "VALUES"
* "Integer" = 3
* "B" = Boys

```python
s3 = pd.Series({"KEYS":"VALUES","Integer":3,"B":"Boys"})
s3
```
